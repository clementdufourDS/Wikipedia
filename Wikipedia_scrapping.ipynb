```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# J'ouvre ma session request
s = requests.Session()
#Je crée une fonction qui va me permettre de créer ma soup avec un url
def get_soup(url):
    response = s.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup
```

```python
# Je crée des listes vides dans lesquelles je vais mettre les éléments que je vais scrapper
countries = []
url = []
flag = []
soup = get_soup('https://fr.wikipedia.org/wiki/Liste_des_pays_du_monde')
#Je vais d'abord scrapper les balises td car c'est là que se trouve les informations que je souhaite scrapper
all_td = soup.find_all("td")

for i in range(len(all_td)):
    if all_td[i].find('img') != None:
        flag.append(all_td[i].find('img')['src'])
        url.append(all_td[i+1].find('a')['href'])
        countries.append(all_td[i+1].find('a').text) 
```

```python
lang = []
cap = []
sup = []
pop = []
for i in url:
    soup = get_soup('https://fr.wikipedia.org'+i)
    # Je vais scrapper toutes les balises td et th car elles contiennent les informations que je veux récupérer
    tr = soup.find_all(['td','th'])
    for i in range(len(tr)):
        if tr[i].find('a',{'title':'Liste des capitales du monde'}) != None:
            lang.append(tr[i-1].find('a').text)
            cap.append(tr[i+1].find('a').text)
        if tr[i].find('a',{'title':'Kilomètre carré'}) != None:
            sup.append(tr[i].text)
        if tr[i].find('a',{'title':'Liste des pays par population'}) != None:
            pop.append(tr[i+1].text)
pop = [i for i in pop if i.startswith('\n')]
```

```python
print(len(cap))
print(len(url))
print(len(flag))
print(len(countries))
print(len(lang))
print(len(pop))
print(len(sup))
```

```python
def rm_let(liste):
    no_let_liste = []
    regroup = []
    for p in liste:
        #J'enléve les '/n' et les espaces
        p = p.strip()
        # Je split chaque élément de ma liste
        for i in p.split():
            try:
                #Si l'élement est un chiffre je le rajoute dans ma liste regroup
                int(i)
                regroup.append(i)
            except:
                #Si je passe sur un élement qui n'est pas un nombre et que ma liste regroup n'est pas vide:
                if len(regroup) > 0:
                    if i[:3].isnumeric() == True:
                    #Si les 2 premiers élements sont des chiffres je les rajoutes à ma liste regroup
                        regroup.append(i[:3])
                    elif i[:2].isnumeric() == True:
                        regroup.append(i[:2])
                    #J'ajoute à ma liste no_let_liste ce que j'ai dans liste regroup que je join
                    no_let_liste.append(' '.join(regroup))
                else:
                    #Si la liste regroup est vide, je rajoute un None à ma liste
                    no_let_liste.append(None)
                # Je vide la liste regroup
                regroup = []
                break
                
    return no_let_liste
```

```python
def clean_num(liste):
    clean_liste = []
    for i in liste:
        s = i.strip().replace(u'\xa0', u'')
        s_clean = ''
        for i in s:
            try:
                int(i)
                s_clean += i
            except:
                clean_liste.append(s_clean)
                break
    return clean_liste
```

```python
popul = rm_let(pop)
```

```python
superfi = rm_let(sup) 
```

```python
df = pd.DataFrame()
df['Pays'] = countries
df['Url'] = url 
df['Drapeaux'] = flag
df['Capitale'] = cap 
df['Langues'] = lang 
df['Population (hab)'] = popul
df['Superficie (km2)'] = superfi
```
